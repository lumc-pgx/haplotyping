```python
cypdat = open('complete_CYPTAM_data.csv','r')

cyptam = {}

line = cypdat.readline()
for line in cypdat:
    record = line.strip().split('\t')
    if record[0] not in cyptam.keys():
        cyptam[record[0]] = {}
        cyptam[record[0]]['genotype'] = record[1]
        cyptam[record[0]]['phenotype']= record[2]
    else:
        print record[0]
        
```

```python
splate = open('cyptam_plate3.csv','r')
seqids = []
for line in splate:
    for record in line.strip().split('\t'):
        seqids.append(record)
```

```python
cnt = 1
for pid in seqids:
    if pid in cyptam.keys():
        print 'lbc'+str(cnt) + '\t' + pid + '\t' + str(cyptam[pid]['genotype']) + '\t' + str(cyptam[pid]['phenotype'])
        cnt = cnt + 1
    else:
        print 'lbc'+str(cnt) + '\t' + pid
        cnt = cnt + 1
```

```python
pacb = open('assignments.txt','r')
pbgen = {}
for line in pacb:
    record = line.strip()
    if record != '':
        bc = record.split('_')[0].split('--')[1]
        if bc not in pbgen.keys():
            pbgen[bc] = []
            pbgen[bc].append(record.split('NumReads')[1])
        else:
            pbgen[bc].append(record.split('NumReads')[1])
```
